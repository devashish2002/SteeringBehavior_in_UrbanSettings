```python
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib.image as mpimg
import imageio
import os
%matplotlib inline
```

```python
data = pd.read_csv('/Users/devashishkhulbe/Desktop/rcnn/416_1.csv')
data.head()
```

```python
data['time'] = pd.to_datetime(data['time'])
times1 = data.time.unique()
```

```python
len(times1)
```

```python
plt.plot(data.X, data.Y, '.', alpha=0.3)
```

```python
int(times1[1]-times1[0])
```

```python
(10*100000000)/float(int(times1[1]-times1[0]))
```

```python
count = 0
prevFrames = []

for time in times1:
    temp = data[data.time == time]
    img = '/Users/devashishkhulbe/Desktop/rcnn/videos/416_op/'+str(count)+'.png'
    img = mpimg.imread(img)
    plt.plot(temp.X, temp.Y, '.', color='blue', alpha=0.5)
    
    for times in prevFrames:
        temp1 = data[data.time == times]
        plt.plot(temp1.X, temp1.Y, '.', color='blue', alpha=(20*100000000)/float(int(time-times)))
        
    prevFrames.append(time)
    plt.imshow(img)
    plt.savefig('/Users/devashishkhulbe/Desktop/rcnn/videos/416_op2/'+str(count)+'.png')
    plt.clf()
    count+=1
    print count
```

```python
png_dir = '/Users/devashishkhulbe/Desktop/rcnn/videos/416_op1/'
images = []
file_names = []

for i in range(len(times1)):
    file_names.append(png_dir+str(i)+'.png')
    
for file in file_names:
    images.append(imageio.imread(file))
    #if file_name.endswith('.png'):
    #    file_path = os.path.join(png_dir, file_name)
    #    images.append(imageio.imread(file_path))     
#file_names.sort(key=lambda x:x.split('.')[0].split('/')[7])
imageio.mimsave('416_10.gif', images, duration=0.05)
```
